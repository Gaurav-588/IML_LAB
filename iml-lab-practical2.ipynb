```python
import numpy as np
```

**Write a NumPy program to implement following operation**

[1] --> to convert a list of numeric values into a one-dimensional NumPy array

```python
arr1 = np.array([1, 2, 3])

print(arr1)
```

[2] --> to create a 3x3 matrix with values ranging from 2 to 10

```python
arr2 = np.arange(2, 11).reshape(3, 3)

print(arr2)
```

[3] --> to append values at the end of an array

```python
arr1 = np.append(arr1, [4])

print(arr1)
```

[4] --> to create another shape from an array without changing its data (3*2 to 2*3)

```python
arr4 = np.arange(0, 6).reshape(3, 2)
print("Array (3 * 2):")
print(arr4)

arr5 = arr4.reshape(2, 3)
print("\nArray (2 * 3):")
print(arr5)
```
