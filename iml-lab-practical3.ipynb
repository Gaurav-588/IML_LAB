```python
import numpy as np
```

**Write a NumPy program to implement following operation**

[1] --> to split an array of 14 elements into 3 arrays, each with 2, 4, and 8 elements in the original order

```python
arr = np.arange(14)

sub_array1, sub_array2, sub_array3 = np.split(arr, [2, 6])

print("Array:")
print(arr)

print("\nArray 1:")
print(sub_array1)

print("\nArray 2:")
print(sub_array2)

print("\nArray 3:")
print(sub_array3)
```

[2] --> to stack arrays horizontally (column wise)

```python
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

stacked_array = np.column_stack((array1, array2))

print("Array Stack (Column-wise):")
print(stacked_array)
```
