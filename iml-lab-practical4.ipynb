```python
import numpy as np
```

**Write a NumPy program to implement following operation**

[1] --> to add, subtract, multiply, divide arguments element-wise

```python
arr = np.array([1, 2, 3])
arr1 = np.array([4, 5, 6])

print("Array 1:")
print(arr)

print("Array 2:")
print(arr1)
print()

sum = np.add(arr, arr1) # arr + arr1
subtract = np.subtract(arr, arr1) # arr - arr1
multiply = np.multiply(arr, arr1) # arr * arr1
divide = np.divide(arr, arr1) # arr / arr1

print("Addition:")
print(sum)

print("Subtraction:")
print(subtract)

print("Multiplication:")
print(multiply)

print("Division:")
print(divide)
```

[2] --> to round elements of the array to the nearest integer

```python
arr3 = np.array([1, 2, 3.5, 4.8, 5, 6.8]).round()

print(arr3)
```

[3] --> to calculate mean across dimension, in a 2D numpy array

```python
arr4 = np.array([[1, 2, 3], [4, 5, 6]])
print("Array:")
print(arr4)
print()

mean_array = np.mean(arr4)
mean_array1 = np.mean(arr4, axis=0)
mean_array2 = np.mean(arr4, axis=1)

print("Mean Array 1:")
print(mean_array)

print("Mean Array 2:")
print(mean_array1)

print("Mean Array 3:")
print(mean_array2)
```

[3] --> to calculate the difference between neighboring elements, element-wise of a given array

```python
arr5 = np.array([[1, 2, 3], [4, 5, 6]])
array_diff = np.diff(arr5)
print("Diffrence:")
print(array_diff)
```
