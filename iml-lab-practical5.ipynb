```python
import numpy as np
```

**Write a NumPy program to implement following operation**

[1] --> to find the maximum and minimum value of a given flattened array

```python
arr = np.array([1, 2, 3, 4, 5, 6])

print("Array:")
print(arr)

print("\nMinimum:", arr.min())
print("\nMaximum:", arr.max())
```

[2] --> to compute the mean, standard deviation, and variance of a given array along the second axis

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print("Array:")
print(arr2)

mean_array1 = np.mean(arr2, axis=0)
std_array1 = np.std(arr2, axis=0)
var_array1 = np.var(arr2, axis=0)

print("\nMean with axis 0:")
print(mean_array1)

print("\nStandard deviation with axis 0:")
print(std_array1)

print("\nVariance with axis 0:")
print(var_array1)

mean_array2 = np.mean(arr2, axis=1)
std_array2 = np.std(arr2, axis=1)
var_array2 = np.var(arr2, axis=1)

print("\nMean with axis 1:")
print(mean_array2)

print("\nStandard deviation with axis 1:")
print(std_array2)

print("\nVariance with axis 1:")
print(var_array2)
```
