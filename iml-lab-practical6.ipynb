**Write a Pandas program to implement following operation**

[1] --> to convert a NumPy array to a Pandas series

```python
arr = np.array([1, 2, 3])
series = pd.Series(arr)

print(series)
```

[2] --> to convert the first column of a DataFrame as a Series

```python
data = {'numbers': [1, 2, 3, 4, 5, 6], 'alphabates': ['a', 'b', 'c', 'd', 'e', 'f']}
df = pd.DataFrame(data)

df_to_series = df['numbers']

print(df_to_series)
```

```python
series1 = pd.Series([1, 3, 2])
print("Sorted Series:", np.sort(series1))
```
