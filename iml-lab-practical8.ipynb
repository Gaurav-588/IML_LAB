**Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.**

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset_path = "/kaggle/input/nifty50-stock-market-data/RELIANCE.csv"
df = pd.read_csv(dataset_path)

df['Date'] = pd.to_datetime(df['Date'])

company_name = "RELIANCE"
start_date = "2020-01-01"
end_date = "2020-12-29"

filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df['Date'], filtered_df['Open'], label='Opening Price', linestyle='-', color='green')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Closing Price', linestyle='-', color='red')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title(f'Stock Prices for {company_name} between {start_date} and {end_date}')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
